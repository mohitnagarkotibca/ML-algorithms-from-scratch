```python
import pandas as pd
from IPython.display import Markdown
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
def prep_data():
    data= pd.read_csv('data/heart.csv')
    for col in data.columns:
        if col != 'target':
            data[col]= data[col]/data[col].mean()
    return data
data= prep_data()
def split_data(data):
    X= data.drop('target',axis=1).copy()
    Y= data['target']
    return train_test_split(X,Y)
train_x,val_x,train_y,val_y= split_data(data)
```

```python
%xmode plain
class logistic_classification:
    def __init__(self,alpha,num_iters):
        self.alpha= alpha
        self.num_iters = num_iters
        self.theta= None
        self.bias= None
        self.N= None
        
    def score(self,Y,Y_pred):
        accuracy= accuracy_score(Y_pred,Y)
        return accuracy
            
    def sigmoid(self,z):
        return 1/ (1+np.exp(-z))
    
    def fit(self,X,Y):# X (227, 13)
        self.N= len(Y)
        self.theta= np.zeros(X.shape[1])#(13,)
        self.bias= 0
        h_x= np.dot(X,self.theta) +self.bias  #(227,)
        h_x= self.sigmoid(h_x)
        for _ in range(self.num_iters):
            d_theta= (1/self.N) * (self.alpha * np.dot(X.T, h_x - Y))
            d_bias= (1/self.N) * np.sum(h_x-Y)
            self.theta= self.theta - d_theta
            self.bias= self.bias - d_bias
            h_x= self.sigmoid(np.dot(X,self.theta) +self.bias)
        print('Trained with :',round(self.score(Y,h_x>0.5) *100) ,'% accuracy on training set')
        
    def predict(self,X,val_y):
        answer= self.sigmoid( np.dot(X,self.theta) + self.bias)>0.5
        print('Got Accuracy of :',accuracy_score(answer,val_y)*100,' % On test set')    
        return answer
lc= logistic_classification(0.01,800)
lc.fit(train_x,train_y)
answer= lc.predict(val_x,val_y)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm= confusion_matrix(val_y,answer)
```

```python
print(cm)
cm=np.ravel(cm)
cm
```

```python
precision= cm[3]/(cm[1]+cm[3])
recall= cm[3]/(cm[3]+cm[2])
f1_score= (2*precision*recall)/(precision+recall)
f1_score
```

## Q1: What is the effect on the coefficients of logistic regression if two predictors are highly correlated?

```python
#finding accuracy without high corelated features
data= prep_data()
train_x,val_x,train_y,val_y= split_data(data)
print('Without co related features: ')
# print(data.columns)
lc.fit(train_x,train_y)
lc.predict(val_x,val_y)
lr.fit(train_x,train_y)
a= lr.predict(val_x)
df1= pd.DataFrame([train_x.columns.T,lr.coef_.T]).T

#adding a highly co related feature
data['high_age_corr']= data['age']*0.5+0.4
data['cp_high_corr']= data['cp']*0.5+0.4

#findinf accuracy with corelated features
print('-'*40)
print('with corelated features')
# print(data.columns)
train_x,val_x,train_y,val_y= split_data(data)
lc.fit(train_x,train_y)
lc.predict(val_x,val_y)

lr.fit(train_x,train_y)
a= lr.predict(val_x)
df2= pd.DataFrame([train_x.columns.T,lr.coef_.T]).T
print('-'*40)
print('with more corelated features :')
data['trestbps_corr']= data['trestbps']*0.5+0.4
data['restecg_corr']= data['restecg']*0.5+0.4
# print(data.columns)
train_x,val_x,train_y,val_y= split_data(data)
lr.fit(train_x,train_y)
a= lr.predict(val_x)
lc.fit(train_x,train_y)
lc.predict(val_x,val_y)
df3= pd.DataFrame([train_x.columns.T,lr.coef_.T]).T
```

## Q2: What relationships exist between a logistic regressionâ€™s coefficient and the Odds Ratio? 

In general with any algorithm, coefficient getting assigned to a variable denotes the significance of that particular variable. High coefficient value means the variable is playing a major role in deciding the boundary (in case of logistic).

Odds ratio tells the changes produced in output variable per unit change in that particular input variable.

##  Q4: Why is logistic regression called regression and not classification?

- The main difference between regression and classification is that the output variable in regression is numerical (or continuous) while that for classification is categorical (or discrete).

- Logistic regression is basically a supervised classification algorithm. However, the model builds a regression model just like linear regression to predict the probability.

- it either predict 0 or 1
