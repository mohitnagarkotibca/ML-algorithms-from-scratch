```python
import pandas as pd
from IPython.display import Markdown
from sklearn.model_selection import train_test_split
import numpy as np
data= pd.read_csv('data/heart.csv')
X= data.drop('target',axis=1).copy()
Y= data['target']
train_x,val_x,train_y,val_y= train_test_split(X,Y)
```

```python
class logistic_classification:
    def __init__(self,alpha,num_iters):
        self.alpha= alpha
        self.num_iters = num_iters
        self.theta= None
        self.bias= None
        self.N= None
        
    def score(self,Y,Y_pred):
        accuracy = np.sum(Y == Y_pred) / len(Y)
        return accuracy
            
    def sigmoid(self,z):
        return 1/ (1+np.exp(-z))
    
    def fit(self,X,Y):# X (227, 13)
        self.N= len(Y)
        self.theta= np.zeros(X.shape[1])#(13,)
        self.bias= 0
        h_x= np.dot(X,self.theta) +self.bias  #(227,)
        h_x= self.sigmoid(h_x)
        for _ in range(self.num_iters):
            d_theta= (1/self.N) * (self.alpha * np.dot(X.T, h_x - Y))
            d_bias= (1/self.N) * np.sum(h_x-Y)
            self.theta= self.theta - d_theta
            self.bias= self.bias - d_bias
            h_x= self.sigmoid(np.dot(X,self.theta) +self.bias)
        print('Trained with ',round(self.score(Y,h_x) *100) ,'% accuracy')
        
    def predict(self,X):
        h_x= self.sigmoid( np.dot(X,self.theta) + self.bias)
        return h_x
lc= logistic_classification(0.01,800)
lc.fit(train_x,train_y)
answer= lc.predict(val_x)
print('Got Accuracy of :',round(lc.score(val_y,answer)*100),' % On test set')
```
